```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import joblib

from lightgbm import LGBMRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error


# =========================================================
# 1. 설정
# =========================================================
solar_data_path = './태양광 발전량 예측 폴더/태양광_발전량데이터_24-07-29-11시.csv'
weather_data_path = './태양광 발전량 예측 폴더/태양광_날씨데이터_24-07-29-11시.csv'

model_save_dir = './태양광 발전량 예측 폴더/'

inverters = {
    'YAG_HSM_SWR_0001': [1015642044025, 1015642044102, 1015642044093],
    'GJM_GSG_SGD_0001': [1018282223012, 1018282225009]
}


# =========================================================
# 2. 유틸 함수
# =========================================================
def convert_pcp(value):
    value = str(value)

    if '강수없음' in value:
        return 0.0
    elif '미만' in value:
        num = re.findall(r'\d+', value)
        if num:
            return float(num[0]) - 0.5
        else:
            return 0.0
    else:
        num = re.findall(r'\d+\.\d+|\d+', value)
        if num:
            return float(num[0])
        else:
            return 0.0


def evaluate_regression(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    non_zero_mask = y_true != 0
    if non_zero_mask.sum() > 0:
        mape = np.mean(
            np.abs((y_true[non_zero_mask] - y_pred[non_zero_mask]) / y_true[non_zero_mask])
        ) * 100
    else:
        mape = np.nan

    mbd = np.mean(y_pred - y_true)

    return {
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'mape': mape,
        'mbd': mbd
    }


def add_solar_features(inverter_data):
    df = inverter_data.sort_values('datetime').copy()

    # -----------------------------------------------------
    # 기본 시간 피처
    # -----------------------------------------------------
    df['hour'] = df['datetime'].dt.hour
    df['month'] = df['datetime'].dt.month
    # df['day'] = df['datetime'].dt.day
    df['dayofyear'] = df['datetime'].dt.dayofyear

    # 주기성 피처
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

    df['dayofyear_sin'] = np.sin(2 * np.pi * df['dayofyear'] / 365)
    df['dayofyear_cos'] = np.cos(2 * np.pi * df['dayofyear'] / 365)

    # 저일사 구간
    df['is_low_sun'] = ((df['hour'] <= 8) | (df['hour'] >= 18)).astype(int)

    df['sun_curve'] = np.cos((df['hour'] - 12) * np.pi / 12)
    df['sun_curve'] = df['sun_curve'].clip(lower=0)

    # # 시간대 구간 피처
    # df['is_morning'] = ((df['hour'] >= 6) & (df['hour'] <= 9)).astype(int)
    # df['is_noon'] = ((df['hour'] >= 10) & (df['hour'] <= 14)).astype(int)
    # df['is_afternoon'] = ((df['hour'] >= 15) & (df['hour'] <= 19)).astype(int)

    # -----------------------------------------------------
    # 관측치 기준 lag 피처
    # 주의: 이건 "관측치 n개 전"이지 "n시간 전"은 아닐 수 있음
    # -----------------------------------------------------
    df['yield_lag1'] = df['DYield_diff'].shift(1)
    df['yield_lag2'] = df['DYield_diff'].shift(2)
    # df['yield_lag3'] = df['DYield_diff'].shift(3)
    df['yield_lag6'] = df['DYield_diff'].shift(6)
    # df['yield_lag24obs'] = df['DYield_diff'].shift(24)
    # df['yield_lag48obs'] = df['DYield_diff'].shift(48)

    df['pac_lag1'] = df['Pac'].shift(1)
    df['pac_lag2'] = df['Pac'].shift(2)
    # df['pac_lag3'] = df['Pac'].shift(3)
    df['pac_lag6'] = df['Pac'].shift(6)
    # df['pac_lag24obs'] = df['Pac'].shift(24)

    # -----------------------------------------------------
    # 진짜 시간 기준 lag 피처
    # datetime 기준 self-merge
    # -----------------------------------------------------
    base = df[['datetime', 'DYield_diff', 'Pac', 't1h', 'reh', 'wsd', 'sky', 
            #    'rn1', 'rn1v'
               ]].copy()

    # lag_1h = base.copy()
    # lag_1h['datetime'] = lag_1h['datetime'] + pd.Timedelta(hours=1)
    # lag_1h = lag_1h.rename(columns={
    #     'DYield_diff': 'yield_lag_1h',
    #     'Pac': 'pac_lag_1h',
    #     't1h': 't1h_lag_1h',
    #     'reh': 'reh_lag_1h',
    #     'wsd': 'wsd_lag_1h',
    #     'sky': 'sky_lag_1h',
    #     'rn1': 'rn1_lag_1h',
    #     'rn1v': 'rn1v_lag_1h'
    # })

    lag_24h = base.copy()
    lag_24h['datetime'] = lag_24h['datetime'] + pd.Timedelta(hours=24)
    lag_24h = lag_24h.rename(columns={
        'DYield_diff': 'yield_lag_24h',
        'Pac': 'pac_lag_24h',
        't1h': 't1h_lag_24h',
        'reh': 'reh_lag_24h',
        'wsd': 'wsd_lag_24h',
        'sky': 'sky_lag_24h',
        # 'rn1': 'rn1_lag_24h',
        # 'rn1v': 'rn1v_lag_24h'
    })

    # df = df.merge(lag_1h, on='datetime', how='left')
    df = df.merge(lag_24h, on='datetime', how='left')

    # -----------------------------------------------------
    # rolling 통계
    # 현재값 누수 방지를 위해 shift(1) 후 rolling
    # -----------------------------------------------------
    df['yield_roll3'] = df['DYield_diff'].shift(1).rolling(3).mean()
    df['yield_roll6'] = df['DYield_diff'].shift(1).rolling(6).mean()
    # df['yield_roll12'] = df['DYield_diff'].shift(1).rolling(12).mean()

    # df['yield_std3'] = df['DYield_diff'].shift(1).rolling(3).std()
    # df['yield_std6'] = df['DYield_diff'].shift(1).rolling(6).std()
    # df['yield_std12'] = df['DYield_diff'].shift(1).rolling(12).std()

    df['pac_roll3'] = df['Pac'].shift(1).rolling(3).mean()
    # df['pac_roll6'] = df['Pac'].shift(1).rolling(6).mean()

    # 날씨 rolling
    df['t1h_roll3'] = df['t1h'].shift(1).rolling(3).mean()
    df['reh_roll3'] = df['reh'].shift(1).rolling(3).mean()
    df['wsd_roll3'] = df['wsd'].shift(1).rolling(3).mean()
    # df['sky_roll3'] = df['sky'].shift(1).rolling(3).mean()

    # -----------------------------------------------------
    # 변화량 / 비율 피처
    # -----------------------------------------------------
    df['yield_diff1'] = df['yield_lag1'] - df['yield_lag2']
    # df['yield_diff2'] = df['yield_lag2'] - df['yield_lag3']
    # df['yield_diff24obs'] = df['yield_lag1'] - df['yield_lag24obs']
    # df['yield_diff_24h'] = df['yield_lag1'] - df['yield_lag_24h']

    df['pac_diff1'] = df['pac_lag1'] - df['pac_lag2']
    # df['pac_diff2'] = df['pac_lag2'] - df['pac_lag3']
    # df['pac_diff_24h'] = df['pac_lag1'] - df['pac_lag_24h']

    # df['yield_ratio1'] = df['yield_lag1'] / (df['yield_lag2'] + 1e-6)
    # df['yield_ratio24obs'] = df['yield_lag1'] / (df['yield_lag24obs'] + 1e-6)
    # df['yield_ratio_24h'] = df['yield_lag1'] / (df['yield_lag_24h'] + 1e-6)

    # df['pac_ratio1'] = df['pac_lag1'] / (df['pac_lag2'] + 1e-6)
    # df['pac_ratio_24h'] = df['pac_lag1'] / (df['pac_lag_24h'] + 1e-6)

    # -----------------------------------------------------
    # 날씨 interaction 피처
    # -----------------------------------------------------
    # df['temp_hour'] = df['t1h'] * df['hour']
    df['temp_sq'] = df['t1h'] ** 2
    # df['reh_sq'] = df['reh'] ** 2
    # df['wsd_sq'] = df['wsd'] ** 2

    df['temp_reh'] = df['t1h'] * df['reh']
    # df['temp_wsd'] = df['t1h'] * df['wsd']
    df['sky_reh'] = df['sky'] * df['reh']

    # df['hour_sky'] = df['hour'] * df['sky']
    # df['hour_temp'] = df['hour'] * df['t1h']
    # df['temp_dayofyear'] = df['t1h'] * df['dayofyear_sin']

    # -----------------------------------------------------
    # 추세 피처
    # -----------------------------------------------------
    df['yield_trend_3_6'] = df['yield_roll3'] - df['yield_roll6']
    # df['pac_trend_3_6'] = df['pac_roll3'] - df['pac_roll6']

    # 결측 제거
    df = df.dropna().reset_index(drop=True)

    return df


# =========================================================
# 3. 데이터 로드
# =========================================================
solar_df = pd.read_csv(solar_data_path)
weather_df = pd.read_csv(weather_data_path)


# =========================================================
# 4. 태양광 데이터 전처리
# =========================================================
solar_df['crt_dttm'] = pd.to_datetime(solar_df['crt_dttm'], errors='coerce')
solar_df = solar_df.dropna(subset=['crt_dttm']).copy()

solar_df['datetime'] = solar_df['crt_dttm'].dt.floor('h')
solar_df['date'] = solar_df['crt_dttm'].dt.date

# 시간별 마지막 관측값 선택
hourly_solar = (
    solar_df
    .sort_values(by=['crt_dttm'])
    .groupby(['PowerPlantID', 'SN', 'datetime'])
    .last()
    .reset_index()
)

# 시간당 발전량 계산
hourly_solar['date'] = hourly_solar['datetime'].dt.date
hourly_solar['DYield_diff'] = (
    hourly_solar
    .groupby(['PowerPlantID', 'SN', 'date'])['DYield']
    .diff()
    .fillna(hourly_solar['DYield'])
)

# 음수 제거
hourly_solar['DYield_diff'] = hourly_solar['DYield_diff'].clip(lower=0)

# 이상치 상단 클리핑
upper_limit = hourly_solar['DYield_diff'].quantile(0.995)
hourly_solar['DYield_diff'] = hourly_solar['DYield_diff'].clip(upper=upper_limit)


# =========================================================
# 5. 날씨 데이터 전처리
# =========================================================
weather_df['base_time'] = weather_df['base_time'].apply(lambda x: f"{int(x):04d}")
weather_df['datetime'] = pd.to_datetime(
    weather_df['base_date'].astype(str) + weather_df['base_time'],
    format='%Y%m%d%H%M',
    errors='coerce'
)
weather_df = weather_df.dropna(subset=['datetime']).copy()
weather_df['datetime'] = weather_df['datetime'].dt.floor('h')

# 필요한 컬럼만 선택
solar_selected = hourly_solar[
    ['PowerPlantID', 'SN', 'datetime', 'DYield_diff', 'crt_dttm', 'Pac']
].copy()

weather_selected = weather_df[
    ['solarpower_sn', 'datetime', 'rn1', 'sky', 't1h', 'reh', 'rn1v', 'wsd']
].copy()

# 병합
merged_df = pd.merge(
    solar_selected,
    weather_selected,
    left_on=['PowerPlantID', 'datetime'],
    right_on=['solarpower_sn', 'datetime'],
    how='inner'
)

# 측정 시각이 xx:55 근처라서 floor 후 +1시간 라벨 보정
merged_df['datetime'] = merged_df['datetime'] + pd.Timedelta(hours=1)


# =========================================================
# 6. 타입 정리
# =========================================================
merged_df['rn1'] = merged_df['rn1'].apply(convert_pcp)

merged_df['rn1v'] = merged_df['rn1v'].replace('\\N', np.nan)
merged_df['rn1v'] = pd.to_numeric(merged_df['rn1v'], errors='coerce').fillna(0)

numeric_cols = ['sky', 't1h', 'reh', 'wsd', 'Pac']
for col in numeric_cols:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce')

merged_df = merged_df.dropna(subset=['sky', 't1h', 'reh', 'wsd', 'Pac']).copy()


# =========================================================
# 7. 학습
# =========================================================
results_dict = {}
metrics_summary = []
feature_importance_dict = {}

for plant_id, sn_list in inverters.items():
    for sn in sn_list:
        inverter_data = merged_df[
            (merged_df['PowerPlantID'] == plant_id) &
            (merged_df['SN'] == sn)
        ].copy()

        inverter_data = add_solar_features(inverter_data)

        if len(inverter_data) < 80:
            print(f"Not enough data for PowerPlantID: {plant_id}, SN: {sn}. Skipping. Sample: {len(inverter_data)}")
            continue

        feature_cols = [
            # 기본 날씨
            'rn1', 'sky', 't1h', 'reh', 'rn1v', 'wsd',

            # 시간 주기성
            'hour_sin', 'hour_cos',
            'month_sin', 'month_cos',
            'dayofyear_sin', 'dayofyear_cos',
            'is_low_sun',

            # 핵심 발전 lag
            'yield_lag1', 'yield_lag2', 'yield_lag6',
            'pac_lag1', 'pac_lag2', 'pac_lag6',

            # 진짜 전일 동일시각
            'yield_lag_24h',
            'pac_lag_24h',
            't1h_lag_24h',
            'reh_lag_24h',
            'wsd_lag_24h',
            'sky_lag_24h',

            # rolling
            'yield_roll3', 'yield_roll6',
            'pac_roll3',
            't1h_roll3', 'reh_roll3', 'wsd_roll3',

            # 변화량
            'yield_diff1',
            'pac_diff1',

            # interaction
            'temp_sq',
            'temp_reh',
            'sky_reh',

            # 추세
            'yield_trend_3_6',

            'sun_curve'
        ]

        X = inverter_data[feature_cols].copy()
        y = inverter_data['DYield_diff'].copy()

        # 시간순 분할
        split_idx = int(len(inverter_data) * 0.8)

        train_data = inverter_data.iloc[:split_idx].copy()
        test_data = inverter_data.iloc[split_idx:].copy()

        X_train = train_data[feature_cols].copy()
        y_train = train_data['DYield_diff'].copy()

        X_test = test_data[feature_cols].copy()
        y_test = test_data['DYield_diff'].copy()

        if len(X_train) == 0 or len(X_test) == 0:
            print(f"Train/Test split failed for PowerPlantID: {plant_id}, SN: {sn}")
            continue

        # LightGBM 모델
        model = LGBMRegressor(
                    n_estimators=300,
                    learning_rate=0.03,
                    num_leaves=31,
                    min_data_in_leaf=30,
                    max_depth=8,
                    subsample=0.8,
                    colsample_bytree=0.8,
                    reg_alpha=0.5,
                    reg_lambda=0.5,
                    random_state=42,
                    n_jobs=-1
                )

        model.fit(
            X_train,
            y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            eval_metric='l2'
        )

        # 예측
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # 모델 저장
        model_filename = f'{model_save_dir}model_lgbm_{plant_id}_{sn}.joblib'
        # joblib.dump(model, model_filename)

        # 평가
        train_metrics = evaluate_regression(y_train.values, y_train_pred)
        test_metrics = evaluate_regression(y_test.values, y_test_pred)

        r2_gap = train_metrics['r2'] - test_metrics['r2']
        rmse_gap = test_metrics['rmse'] - train_metrics['rmse']

        print(f'\n====================================================')
        print(f'PowerPlantID: {plant_id}, SN: {sn}')
        print('-------------------- Train Metrics --------------------')
        print(f"MSE  : {train_metrics['mse']}")
        print(f"RMSE : {train_metrics['rmse']}")
        print(f"MAE  : {train_metrics['mae']}")
        print(f"R2   : {train_metrics['r2']}")
        print(f"MAPE : {train_metrics['mape']}%")
        print(f"MBD  : {train_metrics['mbd']}")

        print('-------------------- Test Metrics ---------------------')
        print(f"MSE  : {test_metrics['mse']}")
        print(f"RMSE : {test_metrics['rmse']}")
        print(f"MAE  : {test_metrics['mae']}")
        print(f"R2   : {test_metrics['r2']}")
        print(f"MAPE : {test_metrics['mape']}%")
        print(f"MBD  : {test_metrics['mbd']}")

        print('----------------- Overfitting Check -------------------')
        print(f"R2 Gap   (Train - Test): {r2_gap}")
        print(f"RMSE Gap (Test - Train): {rmse_gap}")

        if r2_gap > 0.15 and rmse_gap > 1.0:
            print("Warning: 과적합 가능성 있음")
        else:
            print("과적합 징후가 아주 크진 않음")

        metrics_summary.append({
            'PowerPlantID': plant_id,
            'SN': sn,

            'Train_MSE': train_metrics['mse'],
            'Train_RMSE': train_metrics['rmse'],
            'Train_MAE': train_metrics['mae'],
            'Train_R2': train_metrics['r2'],
            'Train_MAPE': train_metrics['mape'],
            'Train_MBD': train_metrics['mbd'],

            'Test_MSE': test_metrics['mse'],
            'Test_RMSE': test_metrics['rmse'],
            'Test_MAE': test_metrics['mae'],
            'Test_R2': test_metrics['r2'],
            'Test_MAPE': test_metrics['mape'],
            'Test_MBD': test_metrics['mbd'],

            'R2_Gap': r2_gap,
            'RMSE_Gap': rmse_gap
        })

        # 결과 저장
        results_df = pd.DataFrame({
            'DateTime': test_data['datetime'].values,
            'PowerPlantID': test_data['PowerPlantID'].values,
            'SN': test_data['SN'].values,
            'crt_dttm': test_data['crt_dttm'].values,
            'Actual': y_test.values,
            'Predicted': y_test_pred
        }).reset_index(drop=True)

        results_df['Difference'] = results_df['Actual'] - results_df['Predicted']
        results_df['Accuracy'] = np.where(
            results_df['Actual'] == 0,
            0,
            (1 - np.abs(results_df['Difference']) / np.abs(results_df['Actual'])) * 100
        )

        results_dict[f'{plant_id}_{sn}'] = results_df

        # feature importance
        importance_df = pd.DataFrame({
            'feature': X_train.columns,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)

        feature_importance_dict[f'{plant_id}_{sn}'] = importance_df

        print('\nFeature Importance Top 30:')
        print(importance_df.head(30))

        # 시간별 평균 비교
        hour_summary = (
            results_df
            .assign(hour=pd.to_datetime(results_df['DateTime']).dt.hour)
            .groupby('hour')[['Actual', 'Predicted']]
            .mean()
        )

        print('\nHour Summary:')
        print(hour_summary)

        # -------------------------------------------------
        # 시각화 1: 테스트 구간 Actual vs Predicted
        # -------------------------------------------------
        plot_df = results_df.sort_values('DateTime').reset_index(drop=True)

        plt.figure(figsize=(14, 7))
        plt.plot(plot_df['DateTime'], plot_df['Actual'], label='Actual', marker='o')
        plt.plot(plot_df['DateTime'], plot_df['Predicted'], label='Predicted', marker='o')
        plt.legend()
        plt.xlabel('DateTime')
        plt.ylabel('Hourly Yield')
        plt.title(f'Actual vs Predicted Hourly Yield ({plant_id}, {sn})')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # -------------------------------------------------
        # 시각화 2: Train / Test 예측 비교
        # -------------------------------------------------
        train_plot_df = pd.DataFrame({
            'DateTime': train_data['datetime'].values,
            'Actual': y_train.values,
            'Predicted': y_train_pred
        }).sort_values('DateTime').reset_index(drop=True)

        test_plot_df = pd.DataFrame({
            'DateTime': test_data['datetime'].values,
            'Actual': y_test.values,
            'Predicted': y_test_pred
        }).sort_values('DateTime').reset_index(drop=True)

        plt.figure(figsize=(14, 7))
        plt.plot(train_plot_df['DateTime'], train_plot_df['Actual'], label='Train Actual', alpha=0.7)
        plt.plot(train_plot_df['DateTime'], train_plot_df['Predicted'], label='Train Predicted', alpha=0.7)
        plt.plot(test_plot_df['DateTime'], test_plot_df['Actual'], label='Test Actual', linewidth=2)
        plt.plot(test_plot_df['DateTime'], test_plot_df['Predicted'], label='Test Predicted', linewidth=2)
        plt.legend()
        plt.xlabel('DateTime')
        plt.ylabel('Hourly Yield')
        plt.title(f'Train/Test Prediction Comparison ({plant_id}, {sn})')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # -------------------------------------------------
        # 일 단위 마지막 시점 평가
        # -------------------------------------------------
        daily_df = plot_df.copy()
        daily_df['Date'] = pd.to_datetime(daily_df['DateTime']).dt.date

        last_results_df = (
            daily_df
            .sort_values(by='DateTime')
            .groupby(['PowerPlantID', 'SN', 'Date'])
            .last()
            .reset_index()
        )

        last_results_df = last_results_df.drop(columns=['Date'])
        last_results_df = last_results_df.sort_values(by='DateTime').reset_index(drop=True)

        daily_metrics = evaluate_regression(
            last_results_df['Actual'].values,
            last_results_df['Predicted'].values
        )

        print('\nDaily Metrics')
        print(f"Daily MSE  : {daily_metrics['mse']}")
        print(f"Daily RMSE : {daily_metrics['rmse']}")
        print(f"Daily MAE  : {daily_metrics['mae']}")
        print(f"Daily R2   : {daily_metrics['r2']}")
        print(f"Daily MAPE : {daily_metrics['mape']}%")
        print(f"Daily MBD  : {daily_metrics['mbd']}")

        plt.figure(figsize=(14, 7))
        plt.plot(last_results_df['DateTime'], last_results_df['Actual'], label='Actual', marker='o')
        plt.plot(last_results_df['DateTime'], last_results_df['Predicted'], label='Predicted', marker='o')
        plt.legend()
        plt.xlabel('DateTime')
        plt.ylabel('Hourly Yield')
        plt.title(f'Daily Last Actual vs Predicted Hourly Yield ({plant_id}, {sn})')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()


# =========================================================
# 8. 전체 결과 확인
# =========================================================
print('\n==================== Metrics Summary ====================')
metrics_summary_df = pd.DataFrame(metrics_summary)
print(metrics_summary_df)

for key, df in results_dict.items():
    print(f"\nResults for {key}:")
    print(df.head())
```
