```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import joblib

from lightgbm import LGBMRegressor, early_stopping, log_evaluation
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error


# =========================================================
# 1. 설정
# =========================================================
solar_data_path = './태양광 발전량 예측 폴더/태양광_발전량데이터_24-07-29-11시.csv'
weather_data_path = './태양광 발전량 예측 폴더/태양광_날씨데이터_24-07-29-11시.csv'

model_save_dir = './태양광 발전량 예측 폴더/'

inverters = {
    'YAG_HSM_SWR_0001': [1015642044025, 1015642044102, 1015642044093],
    'GJM_GSG_SGD_0001': [1018282223012, 1018282225009]
}


# =========================================================
# 2. 유틸 함수
# =========================================================
def convert_pcp(value):
    value = str(value)

    if '강수없음' in value:
        return 0.0
    elif '미만' in value:
        num = re.findall(r'\d+', value)
        if num:
            return max(float(num[0]) - 0.5, 0.0)
        return 0.0
    else:
        num = re.findall(r'\d+\.\d+|\d+', value)
        if num:
            return float(num[0])
        return 0.0


def clean_numeric(series, fillna=None):
    series = series.replace('\\N', np.nan)
    series = pd.to_numeric(series, errors='coerce')
    if fillna is not None:
        series = series.fillna(fillna)
    return series


def evaluate_regression(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    non_zero_mask = y_true != 0
    if non_zero_mask.sum() > 0:
        mape = np.mean(
            np.abs((y_true[non_zero_mask] - y_pred[non_zero_mask]) / y_true[non_zero_mask])
        ) * 100
    else:
        mape = np.nan

    mbd = np.mean(y_pred - y_true)

    return {
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'r2': r2,
        'mape': mape,
        'mbd': mbd
    }


# def apply_physical_postprocess(pred, dt_series):
#     pred = np.clip(pred, 0, None)
#     daytime_mask = dt_series.dt.hour.between(6, 20)
#     pred = np.where(daytime_mask, pred, 0.0)
#     return pred

def apply_physical_postprocess(pred, dt_series):
    pred = np.clip(pred, 0, None)

    hour = dt_series.dt.hour
    daytime_mask = hour.between(6, 20)
    pred = np.where(daytime_mask, pred, 0.0)

    # 저출력 시간대 과대예측 완화
    pred = np.where(hour.isin([6, 7, 19, 20]), pred * 0.7, pred)
    pred = np.where(hour.isin([8, 18]), pred * 0.85, pred)

    return pred


# =========================================================
# 3. 피처 엔지니어링
# =========================================================
def add_features(df):
    df = df.sort_values('datetime').copy()

    # -----------------------------------------------------
    # 3-1. 시간 피처
    # -----------------------------------------------------
    df['hour'] = df['datetime'].dt.hour
    df['month'] = df['datetime'].dt.month
    df['dayofyear'] = df['datetime'].dt.dayofyear

    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

    df['dayofyear_sin'] = np.sin(2 * np.pi * df['dayofyear'] / 365)
    df['dayofyear_cos'] = np.cos(2 * np.pi * df['dayofyear'] / 365)

    df['is_low_sun'] = ((df['hour'] <= 8) | (df['hour'] >= 18)).astype(int)
    df['is_daytime'] = ((df['hour'] >= 6) & (df['hour'] <= 19)).astype(int)

    df['sun_curve'] = np.cos((df['hour'] - 12) * np.pi / 12)
    df['sun_curve'] = df['sun_curve'].clip(lower=0)

    # -----------------------------------------------------
    # 3-2. 날씨 파생
    # -----------------------------------------------------
    df['vec_sin'] = np.sin(2 * np.pi * df['vec'] / 360.0)
    df['vec_cos'] = np.cos(2 * np.pi * df['vec'] / 360.0)

    df['wind_u'] = df['uuu']
    df['wind_v'] = df['vvv']

    df['is_rain'] = (df['pty'] > 0).astype(int)
    df['wind_sq'] = df['wsd'] ** 2
    df['cloud_humidity'] = df['sky'] * df['reh']
    df['sun_temp'] = df['sun_curve'] * df['t1h']
    df['sun_sky'] = df['sun_curve'] * df['sky']
    df['sun_wind'] = df['sun_curve'] * df['wsd']
    df['temp_sq'] = df['t1h'] ** 2
    df['temp_reh'] = df['t1h'] * df['reh']
    df['season_temp'] = df['t1h'] * df['dayofyear_sin']
    df['day_rain'] = df['is_daytime'] * df['is_rain']

    # -----------------------------------------------------
    # 3-3. 인버터 전기 / 온도 파생
    # -----------------------------------------------------
    df['pv_power_1'] = df['Upv1'] * df['Ipv1']
    df['pv_power_2'] = df['Upv2'] * df['Ipv2']
    df['pv_power_3'] = df['Upv3'] * df['Ipv3']
    df['pv_power_sum'] = df[['pv_power_1', 'pv_power_2', 'pv_power_3']].sum(axis=1)

    df['temp_diff_mod_amb'] = df['Tmod'] - df['Tamb']
    df['temp_diff_coil_amb'] = df['Tcoil'] - df['Tamb']
    df['temp_diff_mod_coil'] = df['Tmod'] - df['Tcoil']

    df['ac_voltage_mean'] = df[['Uab', 'Ubc', 'Uca']].mean(axis=1)
    df['ac_voltage_std'] = df[['Uab', 'Ubc', 'Uca']].std(axis=1)

    df['ac_current_mean'] = df[['la', 'lb', 'lc']].mean(axis=1)
    df['ac_current_std'] = df[['la', 'lb', 'lc']].std(axis=1)

    # 상태 플래그
    fault_cols = ['Pfault', 'Warm', 'Fault0', 'fault1', 'fault2', 'fault3', 'fault4', 'PVdetection']
    for col in fault_cols:
        df[col] = (pd.to_numeric(df[col], errors='coerce').fillna(0) > 0).astype(int)

    df['fault_count'] = df[fault_cols].sum(axis=1)

    # -----------------------------------------------------
    # 3-4. 발전량 기반 lag
    # -----------------------------------------------------
    df['yield_lag1'] = df['DYield_diff'].shift(1)
    df['yield_lag2'] = df['DYield_diff'].shift(2)
    # df['yield_lag3'] = df['DYield_diff'].shift(3)
    # df['yield_lag6'] = df['DYield_diff'].shift(6)
    df['yield_lag24obs'] = df['DYield_diff'].shift(24)

    df['yield_roll3'] = df['DYield_diff'].shift(1).rolling(3).mean()
    df['yield_roll6'] = df['DYield_diff'].shift(1).rolling(6).mean()
    # df['yield_roll12'] = df['DYield_diff'].shift(1).rolling(12).mean()

    df['yield_diff1'] = df['yield_lag1'] - df['yield_lag2']
    # df['yield_trend_3_6'] = df['yield_roll3'] - df['yield_roll6']

    # -----------------------------------------------------
    # 3-5. 상태값 lag
    # 현재값 누수 줄이기 위해 lag1 위주 사용
    # -----------------------------------------------------
    lag_cols = [
        'Pac', 'Pdc', 'Eff', 
        # 'PF', 'Sac', 'Qac',
        'Tmod', 'Tamb', 
        # 'Tcoil',
        # 'Uab', 'Ubc', 'Uca',
        # 'la', 'lb', 'lc',
        # 'Upv1', 'Ipv1', 'Upv2', 'Ipv2', 'Upv3', 'Ipv3',
        # 'pv_power_1', 'pv_power_2', 'pv_power_3', 
        'pv_power_sum',
        # 'temp_diff_mod_amb', 'temp_diff_coil_amb', 'temp_diff_mod_coil',
        'ac_voltage_mean', 
        # 'ac_voltage_std',
        'ac_current_mean', 
        # 'ac_current_std'
    ]

    for col in lag_cols:
        df[f'{col}_lag1'] = df[col].shift(1)

    # -----------------------------------------------------
    # 3-6. 전일 동일시각 lag
    # -----------------------------------------------------
    base = df[['datetime', 'DYield_diff', 't1h', 
            #    'reh', 'wsd', 
               'sky', 
               'pv_power_sum', 
            #    'Tmod',
               ]].copy()

    lag_24h = base.copy()
    lag_24h['datetime'] = lag_24h['datetime'] + pd.Timedelta(hours=24)
    lag_24h = lag_24h.rename(columns={
        'DYield_diff': 'yield_lag_24h',
        't1h': 't1h_lag_24h',
        # 'reh': 'reh_lag_24h',
        # 'wsd': 'wsd_lag_24h',
        'sky': 'sky_lag_24h',
        'pv_power_sum': 'pv_power_sum_lag_24h',
        # 'Tmod': 'Tmod_lag_24h'
    })
    df = df.merge(lag_24h, on='datetime', how='left')

    # -----------------------------------------------------
    # 3-7. 타깃 생성
    # 현재 시점 X 로 다음 1시간 발전량 예측
    # -----------------------------------------------------
    df['target_yield'] = df['DYield_diff'].shift(-1)

    # -----------------------------------------------------
    # 3-8. 결측 제거
    # -----------------------------------------------------
    df = df.dropna().reset_index(drop=True)

    # 아침/저녁 과대예측 억제 피처
    df['is_very_early'] = (df['hour'] <= 7).astype(int)
    df['is_evening'] = (df['hour'] >= 18).astype(int)
    df['sun_curve_sq'] = df['sun_curve'] ** 2
    df['low_sun_cloud'] = df['is_low_sun'] * df['sky']
    df['low_sun_humidity'] = df['is_low_sun'] * df['reh']

    return df


# =========================================================
# 4. 데이터 로드
# =========================================================
solar_df = pd.read_csv(solar_data_path)
weather_df = pd.read_csv(weather_data_path)


# =========================================================
# 5. 태양광 데이터 전처리
# =========================================================
solar_df['crt_dttm'] = pd.to_datetime(solar_df['crt_dttm'], errors='coerce')
solar_df = solar_df.dropna(subset=['crt_dttm']).copy()

solar_df['datetime'] = solar_df['crt_dttm'].dt.floor('h')
solar_df['date'] = solar_df['crt_dttm'].dt.date

# 시간별 마지막 관측값 선택
hourly_solar = (
    solar_df
    .sort_values(by=['crt_dttm'])
    .groupby(['PowerPlantID', 'SN', 'datetime'])
    .last()
    .reset_index()
)

# 시간당 발전량 계산
hourly_solar['date'] = hourly_solar['datetime'].dt.date
hourly_solar['DYield_diff'] = (
    hourly_solar
    .groupby(['PowerPlantID', 'SN', 'date'])['DYield']
    .diff()
    .fillna(hourly_solar['DYield'])
)

hourly_solar['DYield_diff'] = hourly_solar['DYield_diff'].clip(lower=0)

upper_limit = hourly_solar['DYield_diff'].quantile(0.995)
hourly_solar['DYield_diff'] = hourly_solar['DYield_diff'].clip(upper=upper_limit)


# =========================================================
# 6. 날씨 데이터 전처리
# =========================================================
weather_df['base_time'] = weather_df['base_time'].apply(lambda x: f"{int(x):04d}")
weather_df['datetime'] = pd.to_datetime(
    weather_df['base_date'].astype(str) + weather_df['base_time'],
    format='%Y%m%d%H%M',
    errors='coerce'
)
weather_df = weather_df.dropna(subset=['datetime']).copy()
weather_df['datetime'] = weather_df['datetime'].dt.floor('h')


# =========================================================
# 7. 필요한 컬럼만 선택
# =========================================================
solar_selected = hourly_solar[
    [
        'PowerPlantID', 'SN', 'datetime', 'crt_dttm', 'DYield_diff',
        'Pac', 'Sac', 'Eff', 'PF', 'Pdc',
        'Uab', 'Ubc', 'Uca',
        'la', 'lb', 'lc',
        'Upv1', 'Ipv1', 'Upv2', 'Ipv2', 'Upv3', 'Ipv3',
        'Tmod', 'Tamb', 'Tcoil',
        'Mode', 'Pfault', 'Warm',
        'Fault0', 'fault1', 'fault2', 'fault3', 'fault4',
        'Qac', 'PVdetection'
    ]
].copy()

weather_selected = weather_df[
    ['solarpower_sn', 'datetime', 'pty', 'rn1', 'sky', 't1h', 'reh', 'rn1v', 'uuu', 'vvv', 'vec', 'wsd']
].copy()


# =========================================================
# 8. 병합
# =========================================================
merged_df = pd.merge(
    solar_selected,
    weather_selected,
    left_on=['PowerPlantID', 'datetime'],
    right_on=['solarpower_sn', 'datetime'],
    how='inner'
)

# 측정 시각 보정
merged_df['datetime'] = merged_df['datetime'] + pd.Timedelta(hours=1)


# =========================================================
# 9. 타입 정리
# =========================================================
# 날씨
merged_df['rn1'] = merged_df['rn1'].apply(convert_pcp)

weather_num_zero_cols = ['pty', 'rn1v', 'uuu', 'vvv', 'vec']
for col in weather_num_zero_cols:
    merged_df[col] = clean_numeric(merged_df[col], fillna=0)

weather_num_cols = ['sky', 't1h', 'reh', 'wsd']
for col in weather_num_cols:
    merged_df[col] = clean_numeric(merged_df[col])

# 발전량/상태 숫자 컬럼
solar_num_cols = [
    'Pac', 'Sac', 'Eff', 'PF', 'Pdc',
    'Uab', 'Ubc', 'Uca',
    'la', 'lb', 'lc',
    'Upv1', 'Ipv1', 'Upv2', 'Ipv2', 'Upv3', 'Ipv3',
    'Tmod', 'Tamb', 'Tcoil',
    'Mode', 'Pfault', 'Warm',
    'Fault0', 'fault1', 'fault2', 'fault3', 'fault4',
    'Qac', 'PVdetection'
]
for col in solar_num_cols:
    merged_df[col] = clean_numeric(merged_df[col])

required_cols = ['sky', 't1h', 'reh', 'wsd', 'Pac', 'Pdc', 'Tmod', 'Tamb']
merged_df = merged_df.dropna(subset=required_cols).copy()


# =========================================================
# 10. 학습
# =========================================================
results_dict = {}
metrics_summary = []
feature_importance_dict = {}

for plant_id, sn_list in inverters.items():
    for sn in sn_list:
        inverter_data = merged_df[
            (merged_df['PowerPlantID'] == plant_id) &
            (merged_df['SN'] == sn)
        ].copy()

        inverter_data = add_features(inverter_data)

        if len(inverter_data) < 150:
            print(f"Not enough data for PowerPlantID: {plant_id}, SN: {sn}. Skipping. Sample: {len(inverter_data)}")
            continue

        # feature_cols = [
        #     # 날씨
        #     'pty', 'rn1', 'sky', 't1h', 'reh', 'rn1v', 'wsd',
        #     'wind_u', 'wind_v', 'vec_sin', 'vec_cos',

        #     # 시간
        #     'hour_sin', 'hour_cos',
        #     'month_sin', 'month_cos',
        #     'dayofyear_sin', 'dayofyear_cos',
        #     'is_low_sun', 'is_daytime', 'sun_curve',

        #     # 날씨 interaction
        #     'is_rain', 'wind_sq', 'cloud_humidity',
        #     'sun_temp', 'sun_sky', 'sun_wind',
        #     'temp_sq', 'temp_reh', 'season_temp', 'day_rain',

        #     # 발전량 lag
        #     'yield_lag1', 'yield_lag2', 'yield_lag3', 'yield_lag6', 'yield_lag24obs',
        #     'yield_roll3', 'yield_roll6', 'yield_roll12',
        #     'yield_diff1', 'yield_trend_3_6',
        #     'yield_lag_24h',

        #     # 전일 동일 시각 날씨/상태
        #     't1h_lag_24h', 'reh_lag_24h', 'wsd_lag_24h', 'sky_lag_24h',
        #     'pv_power_sum_lag_24h', 'Tmod_lag_24h',

        #     # 상태 lag
        #     'Pac_lag1', 'Pdc_lag1', 'Eff_lag1', 'PF_lag1', 'Sac_lag1', 'Qac_lag1',
        #     'Tmod_lag1', 'Tamb_lag1', 'Tcoil_lag1',
        #     'Uab_lag1', 'Ubc_lag1', 'Uca_lag1',
        #     'la_lag1', 'lb_lag1', 'lc_lag1',
        #     'Upv1_lag1', 'Ipv1_lag1', 'Upv2_lag1', 'Ipv2_lag1', 'Upv3_lag1', 'Ipv3_lag1',

        #     # 파생 상태 lag
        #     'pv_power_1_lag1', 'pv_power_2_lag1', 'pv_power_3_lag1', 'pv_power_sum_lag1',
        #     'temp_diff_mod_amb_lag1', 'temp_diff_coil_amb_lag1', 'temp_diff_mod_coil_lag1',
        #     'ac_voltage_mean_lag1', 'ac_voltage_std_lag1',
        #     'ac_current_mean_lag1', 'ac_current_std_lag1',

        #     # 고장/상태 플래그
        #     'Pfault', 'Warm', 'Fault0', 'fault1', 'fault2', 'fault3', 'fault4', 'PVdetection',
        #     'fault_count'
        # ]

        feature_cols = [
            # 날씨
            'pty', 'rn1', 'sky', 't1h', 'reh', 'rn1v', 'wsd',
            'wind_u', 'wind_v', 'vec_sin', 'vec_cos',

            # 시간
            'hour_sin', 'hour_cos',
            'month_sin', 'month_cos',
            'dayofyear_sin', 'dayofyear_cos',
            'is_daytime', 'sun_curve',

            # 날씨 interaction
            'is_rain', 'cloud_humidity',
            'sun_temp', 'sun_sky', 'sun_wind',
            'temp_reh', 'season_temp',

            # 발전량 lag
            'yield_lag1', 'yield_lag2', 'yield_lag24obs',
            'yield_roll3', 'yield_roll6',
            'yield_diff1',
            'yield_lag_24h',

            # 전일 동일 시각
            't1h_lag_24h', 'sky_lag_24h', 'pv_power_sum_lag_24h',

            # 상태 lag
            'Pac_lag1', 'Pdc_lag1', 'Eff_lag1',
            'Tmod_lag1', 'Tamb_lag1',
            'pv_power_sum_lag1',
            'ac_voltage_mean_lag1',
            'ac_current_mean_lag1',

            # 상태 플래그
            'Pfault', 'Warm', 'Fault0', 'fault1', 'fault2', 'fault3', 'fault4',
            'PVdetection', 'fault_count',

            # 아침/저녁 과대예측 억제 피처
            'is_very_early', 'is_evening', 'sun_curve_sq',
            'low_sun_cloud', 'low_sun_humidity',
        ]

        X = inverter_data[feature_cols].copy()
        y = inverter_data['target_yield'].copy()

        split_idx = int(len(inverter_data) * 0.8)

        train_data = inverter_data.iloc[:split_idx].copy()
        test_data = inverter_data.iloc[split_idx:].copy()

        X_train = train_data[feature_cols].copy()
        y_train = train_data['target_yield'].copy()

        X_test = test_data[feature_cols].copy()
        y_test = test_data['target_yield'].copy()

        if len(X_train) == 0 or len(X_test) == 0:
            print(f"Train/Test split failed for PowerPlantID: {plant_id}, SN: {sn}")
            continue

        model = LGBMRegressor(
            n_estimators=400,
            learning_rate=0.03,
            num_leaves=31,
            min_data_in_leaf=30,
            max_depth=8,
            subsample=0.8,
            colsample_bytree=0.8,
            reg_alpha=0.5,
            reg_lambda=0.7,
            random_state=42,
            n_jobs=-1
        )
        
        # 규제
        # model = LGBMRegressor(
        #     n_estimators=500,
        #     learning_rate=0.03,
        #     num_leaves=31,
        #     min_data_in_leaf=20,
        #     max_depth=8,
        #     subsample=0.8,
        #     colsample_bytree=0.8,
        #     reg_alpha=0.3,
        #     reg_lambda=0.5,
        #     random_state=42,
        #     n_jobs=-1
        # )

        # 규제 강하게
        # model = LGBMRegressor(
        #     n_estimators=1000,
        #     learning_rate=0.02,
        #     num_leaves=15,
        #     min_data_in_leaf=50,
        #     max_depth=5,
        #     subsample=0.8,
        #     colsample_bytree=0.7,
        #     reg_alpha=1.0,
        #     reg_lambda=1.5,
        #     random_state=42,
        #     n_jobs=-1
        # )

        # model.fit(
        #     X_train,
        #     y_train,
        #     eval_set=[(X_train, y_train), (X_test, y_test)],
        #     eval_metric='l2'
        # )

        model.fit(
            X_train,
            y_train,
            eval_set=[(X_test, y_test)],
            eval_metric='l2',
            callbacks=[
                early_stopping(50),
                log_evaluation(100)
            ]
        )

        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # 다음 1시간 예측이므로 타깃 시각은 현재 시각 + 1h
        train_target_dt = train_data['datetime'] + pd.Timedelta(hours=1)
        test_target_dt = test_data['datetime'] + pd.Timedelta(hours=1)

        y_train_pred = apply_physical_postprocess(y_train_pred, train_target_dt)
        y_test_pred = apply_physical_postprocess(y_test_pred, test_target_dt)

        # =========================
        # 출력 크기별 성능 확인 추가
        # =========================
        low_mask = y_test < 3
        mid_mask = (y_test >= 3) & (y_test < 15)
        high_mask = y_test >= 15

        if low_mask.sum() > 0:
            print("\nLow Output Metrics")
            print(evaluate_regression(y_test[low_mask], y_test_pred[low_mask]))

        if mid_mask.sum() > 0:
            print("\nMid Output Metrics")
            print(evaluate_regression(y_test[mid_mask], y_test_pred[mid_mask]))

        if high_mask.sum() > 0:
            print("\nHigh Output Metrics")
            print(evaluate_regression(y_test[high_mask], y_test_pred[high_mask]))

        # model_filename = f'{model_save_dir}model_lgbm_{plant_id}_{sn}.joblib'
        # joblib.dump(model, model_filename)

        train_metrics = evaluate_regression(y_train.values, y_train_pred)
        test_metrics = evaluate_regression(y_test.values, y_test_pred)

        r2_gap = train_metrics['r2'] - test_metrics['r2']
        rmse_gap = test_metrics['rmse'] - train_metrics['rmse']

        print(f'\n====================================================')
        print(f'PowerPlantID: {plant_id}, SN: {sn}')
        print('-------------------- Train Metrics --------------------')
        print(f"MSE  : {train_metrics['mse']}")
        print(f"RMSE : {train_metrics['rmse']}")
        print(f"MAE  : {train_metrics['mae']}")
        print(f"R2   : {train_metrics['r2']}")
        print(f"MAPE : {train_metrics['mape']}%")
        print(f"MBD  : {train_metrics['mbd']}")

        print('-------------------- Test Metrics ---------------------')
        print(f"MSE  : {test_metrics['mse']}")
        print(f"RMSE : {test_metrics['rmse']}")
        print(f"MAE  : {test_metrics['mae']}")
        print(f"R2   : {test_metrics['r2']}")
        print(f"MAPE : {test_metrics['mape']}%")
        print(f"MBD  : {test_metrics['mbd']}")

        print('----------------- Overfitting Check -------------------')
        print(f"R2 Gap   (Train - Test): {r2_gap}")
        print(f"RMSE Gap (Test - Train): {rmse_gap}")

        if r2_gap > 0.15 and rmse_gap > 1.0:
            print("Warning: 과적합 가능성 있음")
        else:
            print("과적합 징후가 아주 크진 않음")

        metrics_summary.append({
            'PowerPlantID': plant_id,
            'SN': sn,
            'Train_MSE': train_metrics['mse'],
            'Train_RMSE': train_metrics['rmse'],
            'Train_MAE': train_metrics['mae'],
            'Train_R2': train_metrics['r2'],
            'Train_MAPE': train_metrics['mape'],
            'Train_MBD': train_metrics['mbd'],
            'Test_MSE': test_metrics['mse'],
            'Test_RMSE': test_metrics['rmse'],
            'Test_MAE': test_metrics['mae'],
            'Test_R2': test_metrics['r2'],
            'Test_MAPE': test_metrics['mape'],
            'Test_MBD': test_metrics['mbd'],
            'R2_Gap': r2_gap,
            'RMSE_Gap': rmse_gap
        })

        results_df = pd.DataFrame({
            'FeatureTime': test_data['datetime'].values,
            'TargetTime': test_target_dt.values,
            'PowerPlantID': test_data['PowerPlantID'].values,
            'SN': test_data['SN'].values,
            'Actual': y_test.values,
            'Predicted': y_test_pred
        }).reset_index(drop=True)

        results_df['Difference'] = results_df['Actual'] - results_df['Predicted']
        results_df['Accuracy'] = np.where(
            results_df['Actual'] == 0,
            np.nan,
            (np.abs(results_df['Difference']) / np.abs(results_df['Actual'])) * 100
        )

        results_dict[f'{plant_id}_{sn}'] = results_df

        importance_df = pd.DataFrame({
            'feature': X_train.columns,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)

        feature_importance_dict[f'{plant_id}_{sn}'] = importance_df

        print('\nFeature Importance Top 30:')
        print(importance_df.head(30))

        hour_summary = (
            results_df
            .assign(hour=pd.to_datetime(results_df['TargetTime']).dt.hour)
            .groupby('hour')[['Actual', 'Predicted']]
            .mean()
        )

        print('\nHour Summary:')
        print(hour_summary)

        # 시각화
        plot_df = results_df.sort_values('TargetTime').reset_index(drop=True)

        plt.figure(figsize=(14, 7))
        plt.plot(plot_df['TargetTime'], plot_df['Actual'], label='Actual', marker='o')
        plt.plot(plot_df['TargetTime'], plot_df['Predicted'], label='Predicted', marker='o')
        plt.legend()
        plt.xlabel('TargetTime')
        plt.ylabel('Next Hour Yield')
        plt.title(f'Actual vs Predicted Next-Hour Yield ({plant_id}, {sn})')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()


# =========================================================
# 11. 전체 결과 확인
# =========================================================
print('\n==================== Metrics Summary ====================')
metrics_summary_df = pd.DataFrame(metrics_summary)
print(metrics_summary_df)

for key, df in results_dict.items():
    print(f"\nResults for {key}:")
    print(df.head())
```
